## Exemple 1 : Problème de Monty Hall

- Il y a 3 portes. Derrière l'une d'elle se cache une voiture, derrière les deux autres se cachent des chèvres.
 - Le présentateur vous demande une première fois sur quelle porte vous souhaitez parier.
 - Vous lui répondez
 - Ensuite, il élimine une des deux portes que vous n'avez pas choisi (toujours une derrière laquelle une chèvre se trouve).
 - Enfin, il vous propose de changer votre choix pour l'autre porte restante ou de rester sur votre première décision.


```python
import numpy as np
import random


class HumanUser:
    """L'humain peut jouer avec cette classe"""
    def __init__(self):
        pass

    def make_first_choice(self, choices):
        self.first_choice = None
        # Pour être certain que l'utilisateur
        while self.first_choice not in choices:
            print(f"Please enter a value in this list : {choices}")
            self.first_choice = input()
        return self.first_choice
    
    def make_second_choice(self, choices):
        self.second_choice = None
        # Pour être certain que l'utilisateur
        while self.second_choice not in choices:
            print(f"Please enter a value in this list : {choices}")
            self.second_choice = input()
        return self.second_choice

    
class IaSmart:
    """This IA always changes its choice"""
    def __init__(self):
        pass
    
    def make_first_choice(self, choices):
        """Random first choice"""
        self.first_choice = random.choice(choices)
        return self.first_choice
    
    def make_second_choice(self, choices):
        """Change its choice"""
        return [choice for choice in choices if choice != self.first_choice][0]
    
    
class IaDumb:
    def __init__(self):
        """This IA always keeps its choice"""
        pass
    
    def make_first_choice(self, choices):
        """Random first choice"""
        self.first_choice = random.choice(choices)
        return self.first_choice
    
    def make_second_choice(self, choices):
        """Same than first choice"""
        return self.first_choice
        
        
class MontyGame: 
    def __init__(self, player, n_games=1, verbose=False):
        self.score = 0 
        self.player = player
        self.n_games = n_games
        self.verbose = verbose
        
        self.all_doors = ["1", "2", "3"]
        
        self.correct_door = None
        self.excluded_door = None
        self.second_choice_doors = None
        
    def exclude_one_not_correct_door(self):
        """Exclude one of the not selected door"""
        np.random.shuffle(self.all_doors)
        self.excluded_door = [door for door in self.all_doors if ((door != self.correct_door) and (door != self.first_choice))][0]
        self.second_choice_doors = [door for door in self.all_doors if door != self.excluded_door]
        return self.second_choice_doors
        
    def start(self):
        """Initialize by picking randomly the correct door"""
        self.correct_door = random.choice(self.all_doors)
        
    def play_one(self):
        """Run one game"""
        # We choosed a door
        self.start()
        
        if self.verbose:
            print("Just started a new game")

        # First choice
        self.first_choice = self.player.make_first_choice(self.all_doors)
        
        # Exclude an incorrect door
        self.exclude_one_not_correct_door()
        # Second choice
        self.second_choice = self.player.make_second_choice(self.second_choice_doors)
        # If second choice is correct => score increases
        if self.second_choice == self.correct_door:
            self.score += 1
        else:
            pass
        
        if self.verbose & (self.second_choice == self.correct_door):
            print("You won")
        elif self.verbose:
            print("You lost")
        
             
    def play_many(self):
        for _ in range(self.n_games):
            self.play_one()
```

## Human User

```python
player = HumanUser()
game = MontyGame(player=player, n_games=2, verbose=True)
game.play_many()
game.score
```

## IA User

```python
player = IaSmart()
game = MontyGame(player=player, n_games=2, verbose=True)
game.play_many()
game.score
```

```python
player = IaDumb()
game = MontyGame(player=player, n_games=2, verbose=True)
game.play_many()
game.score
```

## Multiple Runs

```python
N_PARTIES = 10

for Ia in [IaSmart, IaDumb]:
    player = Ia()
    game = MontyGame(player=player, n_games=N_PARTIES, verbose=False)
    game.play_many()
    score = game.score
    score_in_percent = score / N_PARTIES
    print(f"{score} parties gagnées sur un total de {N_PARTIES} - {'{:.2f}'.format(score_in_percent)}")
    print("\n")
```

## Statistics & Graphs for Statistics

```python
import pandas as pd
import matplotlib.pyplot as plt

stats_ia = {"IaSmart": [], "IaDumb": []}


for N_PARTIES in range(1, 500):
    for Ia, ia_name in zip([IaSmart, IaDumb], ["IaSmart", "IaDumb"]):
        player = Ia()
        game = MontyGame(player=player, n_games=N_PARTIES, verbose=False)
        game.play_many()
        score = game.score
        score_in_percent = score / N_PARTIES
        score_in_percent = score / N_PARTIES
        stats_ia[ia_name].append(score_in_percent)
        
        
df = pd.DataFrame.from_dict(stats_ia, orient='index').T
df.plot(figsize=(20,10))
plt.axhline(2/3, color='orange', label="proba = 2/3 = 0.6666...")
plt.axhline(1/3, color='blue', label="proba = 1/3 = 0.3333...")
plt.legend(fontsize=15)
plt.title("Pourcentage de victoires en fonction du nombre de partie", fontsize=15)
plt.xlabel("Nombre de parties", fontsize=15)
plt.ylabel("Pourcentage de parties gagnées", fontsize=15)
plt.show()
```

## Exemple 2 : Pierre Feuille Ciseaux

```python
from collections import Counter

class Game:
    def __init__(self, player_1, player_2):
        self.player_1 = player_1
        self.player_2 = player_2
        
        
    def get_choices(self):
        self.choice_1 = self.player_1.make_choice()
        self.choice_2 = self.player_2.make_choice()
        
        if self.choice_1 == self.choice_2:
            self.get_choices()
        else:
            pass

    def determine_winner(self):
        if ((self.choice_1 == "scissor") and (self.choice_2 == "paper")):
            self.winner = self.player_1
            
        elif ((self.choice_1 == "paper") and (self.choice_2 == "scissor")):
            self.winner = self.player_2
            
        elif ((self.choice_1 == "rock") and (self.choice_2 == "paper")):
            self.winner = self.player_2
            
        elif ((self.choice_1 == "paper") and (self.choice_2 == "rock")):
            self.winner = self.player_1
            
        elif ((self.choice_1 == "rock") and (self.choice_2 == "scissor")):
            self.winner = self.player_1
            
        elif ((self.choice_1 == "scissor") and (self.choice_2 == "rock")):
            self.winner = self.player_2
            
        self.looser = self.player_1 if (self.winner == self.player_2) else self.player_2
            
        print(f"{self.winner} has won with {self.winner.choice} against {self.looser} who used {self.looser.choice}")
        
    def run_game(self):
        self.get_choices()
        # On donne les choix de l'adversaire à chaque joueurs
        # Utile seulement pour le SmartPlayer
        self.player_1.opponent_choices.append(self.choice_2)
        self.player_2.opponent_choices.append(self.choice_1)
        self.determine_winner()
            
            
class RandomPlayer:
    def __init__(self, name):
        self.choices = ["rock", "paper", "scissor"]
        self.name = name
        # Choix de l'adversaire (pas utile pour cette classe)
        self.opponent_choices = []
    
    def make_choice(self):
        self.choice = random.choice(self.choices)
        return self.choice
    
    def __repr__(self):
        # Pour les print
        return self.name 
    
    
class ScissorPlayer:
    def __init__(self, name):
        self.choices = ["rock", "paper", "scissor"]
        self.name = name
        # Choix de l'adversaire (pas utile pour cette classe)
        self.opponent_choices = []
    
    def make_choice(self):
        self.choice = "scissor"
        return self.choice
    
    def __repr__(self):
        # Pour les print
        return self.name
    
    
    
class SmartPlayer:
    def __init__(self, name):
        self.choices = ["rock", "paper", "scissor"]
        self.name = name
        # Choix de l'adversaire (utile pour cette classe)
        self.opponent_choices = []
    
    def make_choice(self):
        # Si c'est la première partie, on joue au hasard
        if self.opponent_choices == []:
            self.choice = random.choice(self.choices)
        # Sinon, on a déjà fait au moins une partie
        else:
            # On regarde ce que l'adversaire joue le plus : 
            most_played = Counter(self.opponent_choices).most_common()[0][0]
            # Et on adapte la réponse faisant le choix qui bat celui que l'adversaire 
            # utilise le plus souvent
            if most_played == "scissor":
                self.choice = "rock"
            elif most_played == "rock":
                self.choice = "paper"
            elif most_played == "paper":
                self.choice = "scissor"
        return self.choice
    
    def __repr__(self):
        # Pour les print
        return self.name 
```
