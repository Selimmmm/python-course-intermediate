# Projet Python

## But 
***
Implémenter un programme architecturé en 3 composantes distinctes : 

- Récupération de données via protocole HTTP (scraping ou API)
- Processing des données
- Envoie / alerte via mail, notification, BDD SQL ou NoSQL, API propriétaire (Discord, Google, Slack...), fichiers html / markdown...


## Structure du dossier
***
Le dossier doit respecter la structure du dossier `projet_example` de ce repository, et contient donc <b>a minima </b>: 
- name_of_class_1.py
- name_of_class_2.py
- name_of_class_2.py
- main.py
- requirements.txt
- runtime.txt
- explications.txt

Et tout autre fichier ou dossier que <b>vous jugerez pertinent</b>.


### Fichiers `<name_of_class>`.py
Chaque fichier ne contient que la définition d'une classe utilisée dans le projet


### Fichier main.py
Ce fichier contient la définition du script principal et fait l'usage de la syntaxe `if __name__ == "__main__"`


### Fichier requirements.txt
Ce fichier contient la liste des librairies et leurs versions (seulement pour les librairies non built-in, pour afficher la version de `requests`, on utilise : `pip3 show requests`)


### Fichier runtime.txt
Contient la version de Python utilisé

## Exécution
***
**Vous devez respecter les consignes suivantes**

**Le correcteur n'a besoin de faire que deux choses :**
- **Installer les versions des librairies spécifiées dans `requirements.txt`**
- **Taper la commande `python3 main.py` à la racine de votre dossier**

**Votre code est commenté**
- **En particulier, toute ligne non évidente est commentée**

**Vous respectez a minima les conventions d'écritures suivantes**
- **Variables écrites en snake_case**
- **Noms des fichiers écrites en snake_case**
- **Noms des classes écrites en CamelCase**
